Análisis de casos nulos

```python
#Porcentaje de casos nulos
#Tomando en cuenta eliminar los nan siempre y cuando no sea más del 10% de la información
data_nan = data_train.isnull().sum()
data_nan = data_nan[data_nan > 0]
porcentaje_nan = (data_nan / len(data_train)) * 100
porcentaje_nan
```

```python
columns_to_clean = ['X6', 'X10', 'X78', 'X86', 'X87', 'X73']
data_train_clean = data_train.dropna(subset=columns_to_clean)
data_train_clean.shape
```

Análisis de valores nulos

```python
#Porcentaje de casos nulos a imputar
#Tomando en cuenta eliminar los nan siempre y cuando no sea más del 10% de la información
data_nan2 = data_train_clean.isnull().sum()
data_nan2 = data_nan2[data_nan2 > 0]
porcentaje_nan2 = (data_nan2 / len(data_train_clean)) * 100
porcentaje_nan2
```

Imputación simple para valores nulos

```python
#Imputación de datos faltantes
from sklearn.impute import SimpleImputer
columns_to_imputation = ['X18', 'X19', 'X21']

imputer = SimpleImputer(strategy='mean')
data_train_clean[columns_to_imputation] = imputer.fit_transform(data_train_clean[columns_to_imputation])
data_train_clean.shape
```

#Detección simple de outliers:
Rango intercuartilico

```python
import numpy as np

# Iterar sobre cada variable numérica en la lista
for col in numeric_cols_clean:
    # Calcular el rango intercuartil (IQR)
    q1 = np.percentile(data_train_clean[col], 25)
    q3 = np.percentile(data_train_clean[col], 75)
    iqr = q3 - q1
    
    # Definir los límites inferior y superior
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    # Reemplazar los valores atípicos con la media de la variable
    data_train_clean[col][data_train_clean[col] < lower_bound] = data_train_clean[col].mean()
    data_train_clean[col][data_train_clean[col] > upper_bound] = data_train_clean[col].mean()
```

Análisis PCA

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df = data_train_clean[numeric_cols_clean]
```

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,len(pca.explained_variance_ratio_)+1), pca.explained_variance_ratio_.cumsum(), marker='o', linestyle='--')
plt.title('Varianza explicada por los componentes principales')
plt.xlabel('Número de componentes')
plt.ylabel('Varianza explicada acumulada')
plt.show()
```

```python
pd.set_option('display.max_rows', None)
cumulative_var_ratio = np.cumsum(pca.explained_variance_ratio_)
var_table = pd.DataFrame({'Número de componentes': range(1, len(cumulative_var_ratio)+1),
                          'Varianza explicada acumulada': cumulative_var_ratio})
var_table.head(25)
```

```python
#Se escogen 21 componentes que explican el 80% de la varianza acumulada 
pca = PCA(n_components=21)
df_pca = pca.fit_transform(df_scaled)
df_pca = pd.DataFrame(df_pca, columns=['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10','PC11','PC12','PC13','PC14','PC15','PC16','PC17','PC18','PC19','PC20','PC21'])
```

```python
# Function to give best model score and parameters
def best_model(model):
    print(model.best_score_)    
    print(model.best_params_)
    print(model.best_estimator_)
def get_auc_scores(y_actual, method,method2):
    auc_score = roc_auc_score(y_actual, method); 
    fpr_df, tpr_df, _ = roc_curve(y_actual, method2); 
    return (auc_score, fpr_df, tpr_df)
```
